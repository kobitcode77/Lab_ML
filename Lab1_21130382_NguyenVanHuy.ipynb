# Task 1.1

```python

n = int(input("Nhập vào số nguyên n: "))
list_Number = []
i = 0
while i < n:
  x = int(input("Nhập vào giá trị x" + str(i + 1) + ": "))
  list_Number.append(x)
  i += 1
print("Giá trị lớn nhất là: ", max(list_Number))
print("Giá trị nhỏ nhất là: ", min(list_Number))
print("Tổng là: " + str(sum(list_Number)))
list_Number.sort()
print(list_Number)
duong = 0
am = 0
for i in list_Number:
  if i > 0:
    duong += 1
  elif i < 0:
    am += 1
print("Danh sách có ", str(duong) , "số dương")
print("Danh sách có ", str(am) , "số âm")
```

# Task 1.2

```python

def compute_S(n):
  result = 0
  temp = 1
  for i in range(1, n + 1):
    temp *= i
    result += temp
  return result
```

```python
compute_S(4)
```

# Task 2.2

```python
# Task 2.2. Develop a NumPy program that constructs an array by repeating.
# Input: [1, 2, 3, 4]

arr = np.arange(1, 5)
print(arr)
arr_double = np.tile(arr, 2)
print(arr_double)
arr_triple = np.tile(arr, 3)
print(arr_triple)
```

```python
# Task 2.3. Develop a NumPy program to replace all elements of the NumPy array that are less than a specified value by this specified value.
# # Original array
arr = np.array([(0.42, 0.48, 0.32), (0.74, 0.58, 0.38), (0.51, 0.34, 0.15)])
print(arr)
print()
print(arr[arr > 0.5])
# Replace all elements of the original array with .5 for values which are greater than .5
arr[arr > 0.5] = 0.5
print()
print(arr)
```
